```python
# break statement in for loop

l = [1, "J", "Data", 4.5]
for i in l:
	if i == "Data":
		break
	print(i)
else:
	print("This will be executed when for loop runs sucessfully without any breaks ")
```

```python
# Continue statement in for loop

l = [1, "J", "Data", 4.5]
for i in l:
	if i == "Data":
		continue
	print(i)
else:
	print("This will be executed when for loop runs successfully without any breaks ")
```

## Problem Statement

```python
# print right triangle

# using while loop
row = 1
while row <= 4:
    col = 1
    while col <= row:
        print("*", end="")
        col += 1
    print()
    row += 1
```

```python
## for using for loop

for i in range(4):
    for j in range(i + 1):
        print("*", end="")
    print()
```

```python
# Square of astrik

for i in range(3):
    for j in range(3):
        print("*", end="")
    print()
```

```python
# Using Multiplication
n=5
for i in range(n):
    print("*" * n)
```

```python
# Hollow Squares

for i in range(5):
    for j in range(5):
        if i==0 or i==4 or j==0 or j==4:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```
