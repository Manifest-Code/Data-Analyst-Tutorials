```python
### Loop statements - It allows you to execute a block of statement repeatedly

# while loop - block of code executed repeatedly until a condition is met

n=7
i=1
while i<n:
    print(i)
    i+=1
```

```python
count=5
while count>0:
    print(count)
    count-=1
```

```python
n = 7
i = 1

while i < n:
    print(i)
    i += 1
else:
    print("This will be executed when while runs successfully without any breaks")
```
