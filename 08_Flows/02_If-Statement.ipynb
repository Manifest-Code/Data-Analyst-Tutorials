 # If syntax
 
 If condition True:
    block of code is executed

ex-

```python
a= 100
	
if a > 0:
    print("a is greater than 0")
```

```python
a= 100
	
if a < 0:
    print("a is less than 0")  # block of code will not executed as it is not satisfying if condition
```

```python
weather = 'Rainy'
if (weather == 'Rainy'):
    print("I will not play football")
```

```python
DS_Course = True
if DS_Course:
    print("I am studying DS course")
```

```python
grade = 8
if (grade >=8):
    print("You have excelllent score")
```

```python
number = 42
if (number % 2 == 0):
   print("The Number is even")
```

```python
number = 42
if (number % 2 == 0) & (number >0):
   print("This is the Number of intrest") 
```

```python
grade= 8
co_crr = True
if (grade >=8 and co_crr==True):
   print("You have excelllent score")
```
