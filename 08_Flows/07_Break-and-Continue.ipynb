```python
### break statement - terminate/ exit loop 
n=7
i=1
while i<n:
    print(i)
    i+=1
    if i==3:
        break
else:
    print("This will be executed when while runs sucessfully without any breaks ")
```

```python
### Continue - skips the iteration - doesn't go forword and restart the loop again

n=7
i=1
while i<n:
    i+=1
    if i==3:
        continue
    print(i)
else:
    print("This will be executed when while runs sucessfully without any breaks ")
```
