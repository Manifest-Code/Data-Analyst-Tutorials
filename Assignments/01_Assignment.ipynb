## Q.2. What is an interpreter in Python?
- In Python, an interpreter is the engine that executes your code directly, line by line, without requiring you to manually compile the
  entire program into machine code first.  
  Compilation (to Bytecode): When you run your .py file, the interpreter first checks the syntax. If it's correct, it compiles the source code into an intermediate, low-level form called Bytecode (stored in .pyc files in the __pycache__ folder).  
  Execution (PVM): The Python Virtual Machine (PVM) then reads the bytecode and converts it into machine-level instructions that your computer's CPU can actually understand and run.

## Q.10. What is the difference between implicit and explicit type casting?

- What is the difference between implicit and explicit type casting:  
  1. Implicit Type Casting:  
     Python automatically converts one data type into another.    
     This typically happens during operations involving different numeric types like adding an integer to a float.  
     No data loss during this process.  
     No special functions needed.  

```python
# example:
a = 5         # integer
b = 10.5      # float
c = a + b
print(c, type(c))
```

   2. Explicit Type Casting: When you use built-in functions to force a variable into a specific type.  
      This is necessary when the interpreter doesn't know how to handle the data automatically.  
      Potential for data loss during this process.  
      Uses functions like int(), str(), etc

```python
# example:
a= 5.7         # float
a = int(a)     # converting float to integer
print(a, type(a))
```

## Q. 12. How does the elif statement work?
- The elif statement is short for "else if." Its purpose is to check for multiple specific conditions one after the other.  
  It only runs if the conditions before it were False.  
  How it works in the "Logic Flow":  
  1. The 'if' statement is checked first.  
  2. If the 'if' is False, Python moves to the first 'elif'
  3. If that 'elif' is True, its code runs, and Python skips all remaining elif and else blocks.
  4. If all 'elif' statements are False, the code moves to the else block (if you have one).
     I

```python
#example:

score = 80

if (score >= 90):
    print("Grade: A")
elif (80<=score<90):
    print("Grade: B")   # This runs!
elif (70<=score<80):
    print("Grade: C")   # This is skipped.
else:
    print("Grade: F")
```

## Q.13. What is the difference between for and while loop?

- In Python, both for and while loops are used to repeat a block of code, but they are used in different scenarios based on whether you know 
  the number of iterations in advance.    
  1. The for Loop (Definite Iteration):  
     A for loop is used when you want to iterate over a sequence (like a list, tuple, string) or a fixed range. It is called "definite" because you usually know how many times it will run before the loop starts.

```python
# example:

# Iterating over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

   2. The while Loop (Indefinite Iteration):  
      A while loop repeats as long as a certain condition remains True.  
      It is called "indefinite" because you might not know exactly how many times the loop will run. It depends on when the condition changes.

```python
#example:
# Running until a condition is met
n = 1
while n <= 3:
    print(n)
    n += 1  # Crucial: changing the condition to avoid an infinite loop
```

## Q.5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each

```python
# List:
List_Fruits = ["apple", "banana", "cherry", "date"]
print(f"Original Values of List are: {List_Fruits}")

# Tuple:
Tuple_Fruits = ("apple", "banana", "cherry", "date")
print(f"Original Values of Tuple are: {Tuple_Fruits}")
```

```python
# List modification:
List_Fruits = ["apple", "banana", "cherry", "date"]
List_Fruits[0] = "Strawberry"
print(f"Modified List is: {List_Fruits}")
```

```python
# Tuple modification attempt: It will raise an error as tuples are immutable
Tuple_Fruits = ("apple", "banana", "cherry", "date")
Tuple_Fruits[0] = "Strawberry"  # This will raise a TypeError
print(f"Modified Tuple is: {Tuple_Fruits}")
```

```python
# Tuple modification attempt: Try and except the error
Tuple_Fruits = ("apple", "banana", "cherry", "date")
print(f"Original Tuple: {Tuple_Fruits}")

try:
    Tuple_Fruits[0] = "Strawberry"
except TypeError as e:
    print(f"Error: {e}")
    print("Failure: You cannot change a tuple element after it is created.")
```

# Q.15 Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
number=int(input("Enter a number: "))

if number<0:
    print("The factorial does not exist from negative numbers.")
elif number==0:
    print("The factorial of 0 is 1.")
else:
    
    falctorial = 1
    numb = number

    while numb>0:
        falctorial = falctorial * numb
        numb -= 1
    print(f"The factorial of {number} is {falctorial}")
```
